```python
import numpy as np
import pandas as pd
import statistics
```

```python
path = '/Users/hugo.paula/Documents/hugo_doc/mimo_emb_fts/Notebooks/metrics_data'
data_sets = ['AEC','KSH','AQI','AQB1','AQB6','AQB12']
#data_sets = ['AEC','KSH','AQI']
#models = ['KFTS','PFTS','VAR','SLSTM','LSTM','GRU','RNN','RF','SVR']
models = ['KFTS','PFTS','VAR','SLSTM','LSTM','GRU','RNN','RF','SVR','SLSTM2']
```

```python
pd.set_option('display.max_rows', None)
```

```python
def clean_dataset(df):
    df = df[['window','nrmse']]
    assert isinstance(df, pd.DataFrame) 
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
def group_df_by_window(df):
    windows = df.groupby("window")
    index = windows.indices
    keys = index.keys()
    return windows, keys 
```

```python
def group_insert(df,ds,model):
    df_result = {
        "ds":[],
        "model":[], 
        "window": [],
        "avg_nrmse": []
    }
    
    df = clean_dataset(df)
    i = 0 
    windows,indexes = group_df_by_window(df)
    
#     #test
#     if ds == 'AQB6' and model == 'VAR':
#         print(indexes)
#         print(len(indexes))
    
    for ix in indexes:
        group = windows.get_group(ix)
        nrmse = round(statistics.mean(group.loc[:,'nrmse']),3)

        df_result["ds"].append(ds)
        df_result["model"].append(model)
        df_result["window"].append(i)
        df_result["avg_nrmse"].append(nrmse)
        i = i + 1
    
    df_result = pd.DataFrame(df_result)
    return df_result
```

```python
def process_files(path,data_sets,models):
    result = {}
    for ds in data_sets:
        for mo in models:
            df = pd.read_csv(path+'/'+ds+'/'+mo+'_'+ds+'.csv', sep=',')
            df_result = group_insert(df,ds,mo)
            
#             #test
#             if ds == 'AQB6' and mo == 'VAR':
#                 print(df_result)
#                 print(statistics.mean(df_result.loc[:,'avg_nrmse']))
            
            result[ds+'_'+mo] = df_result            
    return result 
```

```python
final = process_files(path,data_sets,models)
```

```python
df_result = pd.DataFrame(columns=list(['ds','model','window','avg_nrmse']))
```

```python
for i in final.keys():
    df_result = pd.concat([df_result,final[i]],axis=0) 
    
df_result
```
