```python
def clean_dataset(df):
    #df = df[['window','rmse']]
    df = df[['window','rmse','mae','smape']]
    assert isinstance(df, pd.DataFrame) 
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
def compute_avg_std(df,ds,model):
    df_result = {
        "ds":[],
        "model":[], 
        "rmse_avg": [],
        "rmse_std": [],
        "mae_avg": [],
        "mae_std": [],
        "smape_avg": [],
        "smape_std": []
    }
    
    df = clean_dataset(df)
    rmse = round(statistics.mean(df.loc[:,'rmse']),3)
    mae = round(statistics.mean(df.loc[:,'mae']),3)
    smape = round(statistics.mean(df.loc[:,'smape']),3)
    rmse_std = round(statistics.stdev(df.loc[:,'rmse']),3)
    mae_std = round(statistics.stdev(df.loc[:,'mae']),3)
    smape_std = round(statistics.stdev(df.loc[:,'smape']),3)
    
    df_result["ds"].append(ds)
    df_result["model"].append(model)
    df_result["rmse_avg"].append(rmse)
    df_result["rmse_std"].append(rmse_std)
    df_result["mae_avg"].append(mae)
    df_result["mae_std"].append(mae_std)
    df_result["smape_avg"].append(smape)
    df_result["smape_std"].append(smape_std)
    
    df_result = pd.DataFrame(df_result)
    return df_result
    
```

```python
def insert(df,ds,model):
    df_result = {
        "ds":[],
        "model":[], 
        "window": [],
        "rmse": []
    }
    
    df = clean_dataset(df)
    #i = 0 
    
    for i in range(len(df)): 
        df_result["ds"].append(ds)
        df_result["model"].append(model)
        df_result["window"].append(i)
        df_result["rmse"].append(df.iat[i,1])
        #i = i + 1
        
    
    df_result = pd.DataFrame(df_result)
    return df_result
```

```python
def process_files(path,data_sets,models):
    result = {}
    for ds in data_sets:
        for mo in models:
            df = pd.read_csv(path+'/'+ds+'/'+mo+'.csv', sep=',')
            df_result = insert(df,ds,mo)
            print(df_result)
                        
            result[ds+'_'+mo] = df_result            
    return result 
```

```python
final = process_files(path,data_sets,models)
```

## Average and standard deviation 

```python
def process_files_agv_std(path,data_sets,models):
    result = {}
    for ds in data_sets:
        for mo in models:
            df = pd.read_csv(path+'/'+ds+'/'+mo+'.csv', sep=',')
            df_result = compute_avg_std(df,ds,mo)
            print(df_result)
                        
            result[ds+'_'+mo] = df_result            
    return result 
```

```python
final_avg_std = process_files_agv_std(path,data_sets,models)
```
